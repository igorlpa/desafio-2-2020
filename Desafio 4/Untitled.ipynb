## Dividir treino e teste

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
```

```python
for train_index, test_index in split.split(dataset, dataset.Contratar):
    data_train = dataset.loc[train_index]    
    data_test = dataset.loc[test_index]
```

```python
print(data_train.shape)
print(data_test.shape)
```

## Separando features do target

- já removendo as features que não possuem variação
- não há valor faltante

```python
x_train = data_train.drop(['Possui carro', 'Horas de trabalho padrão', 'Maior de idade', 'Contratar'], axis=1)
x_test = data_test.drop(['Possui carro', 'Horas de trabalho padrão', 'Maior de idade', 'Contratar'], axis=1)
```

```python
y_train = data_train.Contratar
y_test = data_test.Contratar
```

### tratar das variáveis categoricas


```python
cat_cols = list(x_train.select_dtypes(include=['object']).columns)
num_cols = x_train._get_numeric_data().columns
print(cat_cols)
print(num_cols)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

referencias:
- https://medium.com/data-hackers/como-usar-pipelines-no-scikit-learn-1398a4cc6ae9
- https://medium.com/vickdata/a-simple-guide-to-scikit-learn-pipelines-4ac0d974bdcf

```python
cat_transformer = Pipeline(steps=[
    ('encoder', OrdinalEncoder()) , 
    ('one-hot encoder', OneHotEncoder(drop='first'))
])
```

```python
num_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer())
])
```

```python
preprocessor = ColumnTransformer(transformers=[
    ('num', num_transformer, num_cols),
    ('cat', cat_transformer, cat_cols)
])
```

```python
xx  =  pd.DataFrame( preprocessor.fit_transform(x_train) )
```

## Treinamento 

```python
from sklearn.preprocessing import StandardScaler
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier
from sklearn.feature_selection import SelectKBest
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('scaler', StandardScaler()),
    #('select', SelectKBest(k=30)),
    ('clf', LGBMClassifier())
])
```

```python
model.fit(x_train, y_train)
```

```python
pred_train = model.predict(x_train)
```

```python
print(classification_report(y_train, pred_train))
```

## Teste

```python
pred_test = model.predict(x_test)
```

```python
print(classification_report(y_test, pred_test))
```

```python
print(accuracy_score(y_test, pred_test))
```
